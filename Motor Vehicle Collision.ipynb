```python
import pandas as pd
import numpy as np

path ="C:\\Users\\tahalam\\Documents\\DS Project\\"
df = pd.read_csv(path+"Motor_Vehicle_Collisions_-_Crashes.csv")

df.head()
```

```python
df["CRASH DATE"] = df["CRASH DATE"].astype(str)
df["Year"] = df["CRASH DATE"].str.split("/").str[2].map(str.strip)

#df.dropna(subset=['ZIP CODE']).reset_index(drop=True)
df["ZIP CODE"] = df["ZIP CODE"].astype(str)
df["ZIP CODE"] = df["ZIP CODE"].str.split(".").str[0].map(str.strip)

print(len(df))

## only take last 5 year excluding 2022
df = df[df["Year"] != "2012"].reset_index(drop=True)
df = df[df["Year"] != "2013"].reset_index(drop=True)
df = df[df["Year"] != "2014"].reset_index(drop=True)
df = df[df["Year"] != "2015"].reset_index(drop=True)
df = df[df["Year"] != "2016"].reset_index(drop=True)
df = df[df["Year"] != "2022"].reset_index(drop=True)

df = df.reset_index(drop=True)

print(len(df))
```

## Combine Count injured or killed

```python
df["Persons Injured or Killed"] = df["NUMBER OF PERSONS INJURED"] + df["NUMBER OF PERSONS KILLED"]
df["Pedestrians Injured or Killed"] = df["NUMBER OF PEDESTRIANS INJURED"] + df["NUMBER OF PEDESTRIANS KILLED"]
df["Cyclist Injured or Killed"] = df["NUMBER OF CYCLIST INJURED"] + df["NUMBER OF CYCLIST KILLED"]
df["Motorist Injured or Killed"] = df["NUMBER OF MOTORIST INJURED"] + df["NUMBER OF MOTORIST KILLED"]

#df["Persons Injured or Killed"] =df["Persons Injured or Killed"].astype(int).replace('\.0', '', regex=True)
df.head()
```

## Add Neighborhood by Zipcode

```python
hood = pd.read_csv(path+'nyc-zip-codes.csv')
#hood = hood[["Neighborhood","ZipCode"]]

hood = hood.rename(columns={'ZipCode':'ZIP CODE'})


hood["ZIP CODE"] = hood["ZIP CODE"].astype(str)
df = df.merge(hood,how='left',on='ZIP CODE')

df.tail()
```

```python
result = df[["CRASH DATE","BOROUGH","ZIP CODE","Year","Persons Injured or Killed","Pedestrians Injured or Killed","Cyclist Injured or Killed",
            "Motorist Injured or Killed","Neighborhood"]]
```

## Count of vehicle and factors

```python
# top factors and vehciles contributing to the collision

s1 = df.copy()

s2 = s1["CONTRIBUTING FACTOR VEHICLE 1"].value_counts().head(11).to_frame(name="Total Collision")
s2.index.name ='Factor'
s2 = s2.reset_index(drop=False)
s2

s3 = s1["CONTRIBUTING FACTOR VEHICLE 2"].value_counts().to_frame(name="Total Collision")
s3.index.name ='Factor'
s3 = s3.reset_index(drop=False)
s3

s4 = s1["CONTRIBUTING FACTOR VEHICLE 3"].value_counts().to_frame(name="Total Collision")
s4.index.name ='Factor'
s4 = s4.reset_index(drop=False)
```

```python
s2 = s2.merge(s3,how='left',on='Factor')
s2 = s2.merge(s4,how='left',on='Factor')
s2["Number Collision"] = s2["Total Collision_x"] + s2["Total Collision_y"] + s2["Total Collision"]
s2 = s2[s2["Factor"]!="Unspecified"]
s2= s2[["Factor","Number Collision"]]
s2 = s2.reset_index(drop=True)
s2
```

```python
r1 = df.copy()

r2 = r1["VEHICLE TYPE CODE 1"].value_counts().head(11).to_frame(name="Total Collision")
r2.index.name ='Vehicle'
r2 = r2.reset_index(drop=False)
r2

r3 = r1["VEHICLE TYPE CODE 2"].value_counts().to_frame(name="Total Collision")
r3.index.name ='Vehicle'
r3 = r3.reset_index(drop=False)
r3

r4 = r1["VEHICLE TYPE CODE 3"].value_counts().to_frame(name="Total Collision")
r4.index.name ='Vehicle'
r4 = r4.reset_index(drop=False)

r2
```

```python
r2 = r2.merge(r3,how='left',on='Vehicle')
r2 = r2.merge(r4,how='left',on='Vehicle')
r2["Number Collision"] = r2["Total Collision_x"] + r2["Total Collision_y"] + r2["Total Collision"]
r2= r2[["Vehicle","Number Collision"]]
r2 = r2.reset_index(drop=True)
r2
```
